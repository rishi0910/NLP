```python
## Data Cleaning and Preprocessing
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
from nltk.stem import PorterStemmer
ps=PorterStemmer()
```

```python
corpus=[]
for i in range(0,len(messages)):
  review=re.sub('[^a-zA-Z]',' ', messages['email_text'][i])
  review=review.lower()
  review=review.split()
  review= [ps.stem(word) for word in review if word not in stopwords.words('english')]
  review=' '.join(review)
  corpus.append(review)
```

```python
corpus
```

```python
## Create bag of words
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=1000)
X=cv.fit_transform(corpus).toarray()
```

**Making BOW and applying lemmatization**

```python
import pandas as pd
data=pd.read_csv('email_spam_dataset.csv')
```

```python
import re
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
swords=stopwords.words('english')
```

```python
corps=[]
for j in range(0,len(data)):
  reviews=re.sub('[^a-zA-Z]',' ',data['email_text'][j])
  reviews=reviews.lower()
  reviews=reviews.split()
  reviews=[lemmatizer.lemmatize(wrod,pos='v') for wrod in reviews if wrod not in swords]
  reviews=' '.join(reviews)
  corps.append(reviews)
```
