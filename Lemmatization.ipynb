```python
!pip install nltk
import nltk
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer=WordNetLemmatizer()
```

POS
Noun - n
Verb - v
Adjective - a
Adverb - r

```python
words=["plays","playing","eats","eating","running","runs"]
for word in words:
  print(lemmatizer.lemmatize(word,pos='v'))
```

```python
lemmatizer.lemmatize('fairly',pos='v'),lemmatizer.lemmatize('sportingly',pos='r')
```
